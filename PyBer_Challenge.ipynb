### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type

 
total_rides=pyber_data_df.groupby(['type']).count()['ride_id'] 
total_rides
```

```python
# 2. Get the total drivers for each city type
total_driver_city_type=city_data_df.groupby(['type']).sum()['driver_count'] 
total_driver_city_type
```

```python
#  3. Get the total amount of fares for each city type
total_fares_city=pyber_data_df.groupby(['type']).sum()['fare'] 
total_fares_city
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_perride_citytype=pyber_data_df.groupby(['type']).mean()['fare'] 
avg_fare_perride_citytype
```

```python
# 5. Get the average fare per driver for each city type. 
#pyber_data_df.groupby(['type','driver_count'])['fare'].mean() 
avgfare_per_driverforcitytype=total_fares_city/total_driver_city_type 
avgfare_per_driverforcitytype
```

```python
#  6. Create a PyBer summary DataFrame. 
PyBer_df=pd.DataFrame({'Total rides for each city type':total_rides, 
                       'Total drivers for each city type':total_driver_city_type, 
                       'Total amount of fares for each city type':total_fares_city, 
                       'Average fare per ride for each city type':avg_fare_perride_citytype, 
                       'Average fare per driver for each city type':avgfare_per_driverforcitytype}) 
PyBer_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
PyBer_df.index.name = None 
PyBer_df
```

```python
#  8. Format the columns.
PyBer_df['Total amount of fares for each city type']=PyBer_df['Total amount of fares for each city type'].map('${:,.2f}'.format) 
PyBer_df['Average fare per ride for each city type']=PyBer_df['Average fare per ride for each city type'].map('${:,.2f}'.format)
PyBer_df['Average fare per driver for each city type']=PyBer_df['Average fare per driver for each city type'].map('${:,.2f}'.format)    
    
    
    
    
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sumoffares_bydatecitytype_df=pyber_data_df.groupby(['type','date']).sum()['fare'].to_frame()
sumoffares_bydatecitytype_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
sumoffares_bydatecitytype_df=sumoffares_bydatecitytype_df.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
faresum_pivot=sumoffares_bydatecitytype_df.pivot(index='date',columns='type',values='fare') 
faresum_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

datepivottable=faresum_pivot.loc['2019-01-01':'2019-04-29'] 
datepivottable
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index) 
datepivottable.index=pd.to_datetime(datepivottable.index) 
datepivottable.info()
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
datepivottable.info() 
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares_df=datepivottable.resample('W').sum() 
weekly_fares_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

ax=weekly_fares_df.plot(figsize=(20,6)) 
# Add a title 
ax.set_title("Total Fare by City Type")
#  Add a x- and y-axis label.
ax.set_xlabel("Month")
ax.set_ylabel("Fare ($USD)")

plt.show()
```
